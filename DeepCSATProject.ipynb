### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Before Doing any manipulation and Data wrangling task on data set we will make a copy of the original data set
df_copy = df.copy()
````

````python
df_copy.info()
````

````python
#Dropping columns with a very high percentage of missing values
df_copy.drop(['connected_handling_time',
              'Customer_City',
              'Product_category',
              'Item_price',
              'order_date_time',
              'Customer Remarks',
              'Order_id'], axis=1, inplace=True)
````

````python
#Checking the dataset structure after removing unnecessary columns
df_copy.info()
````

````python
#Removing duplicate rows from the dataset
df_copy.drop_duplicates(inplace=True)
````

````python
#Converting date columns into datetime format
df_copy['Issue_reported at'] = pd.to_datetime(df_copy['Issue_reported at'],format ='%d/%m/%Y %H:%M', errors = 'coerce')
df_copy['issue_responded'] = pd.to_datetime(df_copy['issue_responded'],format ='%d/%m/%Y %H:%M', errors = 'coerce')
df_copy['Survey_response_Date'] = pd.to_datetime(df_copy['Survey_response_Date'],format='%d-%b-%y', errors='coerce')
````

````python
#Creating a new feature representing response time in hours
df_copy['response_time_hours'] = (df_copy['issue_responded'] - df_copy['Issue_reported at']).dt.total_seconds() / 3600
````

````python
#Extracting day, month and hour information from issue reported time
df_copy['Issue_reported_day'] = df_copy['Issue_reported at'].dt.day
df_copy['Issue_reported_month'] = df_copy['Issue_reported at'].dt.month
df_copy['Issue_reported_hour'] = df_copy['Issue_reported at'].dt.hour
````

````python
#Removing rows with invalid response time values
df_copy[df_copy['response_time_hours'] < 0]
df_copy = df_copy[df_copy['response_time_hours'] >= 0]
````

````python
# Dropping original datetime columns after extracting useful features
df_copy.drop(['Issue_reported at','issue_responded','Survey_response_Date'], axis=1, inplace=True)
````

#### 1. State Your research hypothesis as a null hypothesis and alternate hypothesis.

#### What all missing value imputation techniques have you used and why did you use those techniques?

Missing values in the dataset were handled based on the type of the variables.
Categorical columns such as Agent_name, channel_name, category, and Agent Shift were checked for missing values. If any missing values were present, they were replaced using the mode (most frequent value) of the respective column.

Using the mode for categorical variables helps preserve the most common category without introducing unrealistic values. This ensures that the dataset remains consistent and suitable for further analysis and modeling.

To handle this, Label Encoding was applied using the LabelEncoder method from the scikit-learn library. Label Encoding assigns a unique numerical value to each category within a column.

This method was chosen because the dataset contains several categorical variables and Label Encoding efficiently converts them into numeric form without significantly increasing the dimensionality of the dataset.

### ML Model - 1

````python
# Initialize model
model_lr = LogisticRegression(max_iter=1000)

# Fit the model
model_lr.fit(X_train, y_train)

# Predict
y_pred_lr = model_lr.predict(X_test)
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

accuracy = accuracy_score(y_test, y_pred_lr)
precision = precision_score(y_test, y_pred_lr, average='weighted')
recall = recall_score(y_test, y_pred_lr, average='weighted')
f1 = f1_score(y_test, y_pred_lr, average='weighted')

print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
````

````python
# Visualizing evaluation Metric Score chart

metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']
scores = [accuracy, precision, recall, f1]

plt.figure(figsize=(6,4))
sns.barplot(x=metrics, y=scores)

plt.title("Logistic Regression Evaluation Metrics")
plt.ylim(0,1)

plt.show()
````

#### 2. Cross- Validation & Hyperparameter Tuning

````python
# Define model
lr = LogisticRegression(max_iter=1000)

# Define hyperparameters to tune
param_grid = {
    'C': [0.01, 0.1, 1, 10],
    'solver': ['lbfgs', 'liblinear']
}

# Grid Search with Cross Validation
grid_search = GridSearchCV(
    estimator=lr,
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

# Fit the model
grid_search.fit(X_train, y_train)

# Best model
best_model = grid_search.best_estimator_

# Predict using optimized model
y_pred_tuned = best_model.predict(X_test)
````

````python
accuracy_tuned = accuracy_score(y_test, y_pred_tuned)
precision_tuned = precision_score(y_test, y_pred_tuned, average='weighted')
recall_tuned = recall_score(y_test, y_pred_tuned, average='weighted')
f1_tuned = f1_score(y_test, y_pred_tuned, average='weighted')

print("Accuracy:", accuracy_tuned)
print("Precision:", precision_tuned)
print("Recall:", recall_tuned)
print("F1 Score:", f1_tuned)
````

````python
metrics = ['Accuracy','Precision','Recall','F1 Score']
scores = [accuracy_tuned, precision_tuned, recall_tuned, f1_tuned]

plt.figure(figsize=(6,4))
sns.barplot(x=metrics, y=scores)

plt.title("Tuned Logistic Regression Performance")
plt.ylim(0,1)

plt.show()
````
